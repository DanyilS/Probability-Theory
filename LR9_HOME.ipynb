```python
import numpy as np
import matplotlib.pyplot as plt
# Class that anables 3D plotting
from mpl_toolkits.mplot3d import Axes3D

# Class of color maps
from matplotlib import cm

# This line makes 3D plots in Jupyter Notebook interactive (uncomment to anable)
# %matplotlib notebook
from scipy.stats import multivariate_normal
```

```python
m = [0, 0]
cov = ((2, 0),(0, 1))
dz = multivariate_normal(m, cov)

x,y = np.arange(-3, 3.01, 0.01), np.arange(-3, 3.01, 0.01)
x, y = np.meshgrid(x, y)
pos = np.dstack((x, y))
z = dz.pdf(pos)

fig = plt.figure(figsize=(20,10))
ax1, ax2 = fig.add_subplot(121, projection='3d'), fig.add_subplot(122)
surf, cont = ax1.plot_surface(x,y,z, cmap=cm.coolwarm), ax2.contourf(x,y,z, cmap=cm.coolwarm)
fig.colorbar(cont)

plt.show()
```
