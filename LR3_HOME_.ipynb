```python
#Task_1_3_12
from itertools import product
import itertools
import numpy as np

list_num = list(range(10))

def calc_theor(c):
    lst = list(itertools.product(list_num, repeat=3))
    counter = 0
    for i in lst:
        if len(set(i)) == c:
            counter += 1
    return counter/(len(lst))


def experiment_100000(c):
    tosses = []
    counter = 0
    for i in range(100000):
        tmp = list(np.random.choice(list_num, size=3))
        tosses.append(tmp)
    for i in tosses:
        if len(set(i)) == c:
            counter += 1
    return counter



list_d = [1,2,3]
for i in list_d:
    res_empir = experiment_100000(i)
    print("-------------               -------------              \n\
Theoretical probability for {} different numbers is {}".format(i, calc_theor(i)))
    print("Empirical probability for {} different numbers is {}".format(i, res_empir/100000))
```

```python
# Task_1_3_14_1
import numpy as np
import itertools


def test_theor_1(n, k, p):
    lst = list(itertools.combinations(range(1, n + 1), k))
    counter = 0
    bool = False
    for i in range(len(lst)):
        for j in range(k):
            if lst[i][j] % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1
    return counter/(len(lst))



def test_empir_1(n, k, p):
    counter = 0
    for _ in range(100000):
        tmp = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in tmp:
            if i % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1

    return counter / 100000


exps = [[25, 5, 3, 4], [25, 10, 3, 4]]
print("_____Task_2.1______")
for i in exps:
    print("Theor probability n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % test_theor_1(i[0], i[1], i[2])))
    print("             Empiric probability is {}".format('%.30f' % test_empir_1(i[0], i[1], i[2])))
    print("---- ---- ---- ---- ---- ---- ---- ---- ---- ----")

```

```python
# Task_1_3_14_2
import numpy as np
import itertools


def test_theor_2(n, k, p,q):
    lst = list(itertools.combinations(range(1, n + 1), k))
    counter = 0
    bool = False
    for i in range(len(lst)):
        for j in range(k):
            if lst[i][j] % p == 0 or lst[i][j] % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1
    return counter/(len(lst))



def test_empir_2(n, k, p,q):
    counter = 0
    for _ in range(100000):
        tmp = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in tmp:
            if i % p == 0 or i % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1

    return counter / 100000


exps = [[25, 5, 3, 4], [25, 10, 3, 4]]
print("_____Task_2.2______")
for i in exps:
    print("Theor probability n={}, k={}, p={}, q={} is {}".format(i[0], i[1], i[2], i[3], '%.30f' % test_theor_2(i[0], i[1], i[2], i[3])))
    print("             Empiric probability is {}".format('%.30f' % test_empir_2(i[0], i[1], i[2], i[3])))
    print("---- ---- ---- ---- ---- ---- ---- ---- ---- ----")

```

```python
# Task 1_4_6(1,2,3)

import itertools
import numpy as np

def calc_probability(combination):
    counter1, counter2, counter3 = 0, 0, 0
    for i in combination:
        if len(i) == len(set(i)): counter1 += 1
        elif len(i)  == len(set(i)) + 1: counter2 += 1
        elif len(i) == len(set(i)) + 2: counter3 += 1
    return counter1 / len(combination), counter2 / len(combination), counter3 / len(combination)



tests = [[8,4],[13,5]]
for i in tests:
    list_of_shoes = [x for x in range(1, i[0] + 1)] * 2
    tmp_combination = list(itertools.combinations(list_of_shoes, r=2 * i[1]))
    res1t, res2t, res3t = calc_probability(tmp_combination)
    ans = []
    for j in range(100000):
        ans.append(np.random.choice(list_of_shoes, size=2 * i[1], replace=False))
    res1e, res2e, res3e = calc_probability(ans)
    ans = [res1t, res1e, res2t, res2e, res3t, res3e]
#     print(ans)
    for k in range(0, len(ans), 2):
        print("Theoretical probability when n={}, k={} is: {}".format(i[0], i[1], '%.10f'%ans[k]))
        print("Empirical probability is: {}".format(ans[k+1]))
        print("----------   ---------------------   ----------")
```

```python
# Task_1_2_15(1,2)
import numpy as np

def calc_t(n):
    logs = []
    tmp = []
    for i in list(range(1,2*n+1)):
        logs.append(np.log(i))
    multi = 1/sum(logs)
    for i in logs:
        tmp.append(i*multi)

    res1 = 0
    res2 = 0
    for i in tmp[1::2]:
        res1 += i
    for j in tmp[::2]:
        res2 += j
    ans = [res1,res2]
    return ans


def calc_e(n):
    res1 = 0
    res2 = 0
    for i in range(100000):
        num = np.random.randint(0, 2*n + 1)
        if num % 2 == 0:
            res1 += 1
        else:
            res2 += 1
    ans = [res1/ 10**5, res2/ 10**5]
    return ans


for n in [10, 25]:
    print( "When n ={} theoretical probability is: even = {}, odd = {}\nEmpirical probability is: even = {}, odd = {}\n------ ------".format(n, calc_t(n)[0], calc_t(n)[1], calc_e(n)[0], calc_e(n)[1]))
```
