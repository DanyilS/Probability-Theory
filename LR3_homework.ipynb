```python
#Task_1_3_12
from itertools import product
import itertools
import numpy as np

list_num = list(range(10))

def calc_theor(c):
    lst = list(itertools.product(list_num, repeat=3))
    counter = 0
    for i in lst:
        if len(set(i)) == c:
            counter += 1
    return counter/(len(lst))


def experiment_100000(c):
    tosses = []
    counter = 0
    for i in range(100000):
        tmp = list(np.random.choice(list_num, size=3))
        tosses.append(tmp)
    for i in tosses:
        if len(set(i)) == c:
            counter += 1
    return counter



list_d = [1,2,3]
for i in list_d:
    res_empir = experiment_100000(i)
    print("-------------               -------------              \n\
Theoretical probability for {} different numbers is {}".format(i, calc_theor(i)))
    print("Empirical probability for {} different numbers is {}".format(i, res_empir/100000))
```

```python
# Task_1_3_14_1
import numpy as np
import itertools


def test_theor_1(n, k, p):
    lst = list(itertools.combinations(range(1, n + 1), k))
    counter = 0
    bool = False
    for i in range(len(lst)):
        for j in range(k):
            if lst[i][j] % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1
    return counter/(len(lst))



def test_empir_1(n, k, p):
    counter = 0
    for _ in range(100000):
        tmp = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in tmp:
            if i % p == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1

    return counter / 100000


exps = [[25, 5, 3, 4], [25, 10, 3, 4]]
print("_____Task_2.1______")
for i in exps:
    print("Theor probability n={}, k={}, p={} is {}".format(i[0], i[1], i[2], '%.30f' % test_theor_1(i[0], i[1], i[2])))
    print("             Empiric probability is {}".format('%.30f' % test_empir_1(i[0], i[1], i[2])))
    print("---- ---- ---- ---- ---- ---- ---- ---- ---- ----")

```

```python
# Task_1_3_14_2
import numpy as np
import itertools


def test_theor_2(n, k, p,q):
    lst = list(itertools.combinations(range(1, n + 1), k))
    counter = 0
    bool = False
    for i in range(len(lst)):
        for j in range(k):
            if lst[i][j] % p == 0 or lst[i][j] % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1
    return counter/(len(lst))



def test_empir_2(n, k, p,q):
    counter = 0
    for _ in range(100000):
        tmp = list(np.random.choice(list(range(1, n + 1)), size=k, replace=False))
        bool = False
        for i in tmp:
            if i % p == 0 or i % q == 0:
                bool = True
            else:
                bool = False
                break
        if bool:
            counter += 1

    return counter / 100000


exps = [[25, 5, 3, 4], [25, 10, 3, 4]]
print("_____Task_2.2______")
for i in exps:
    print("Theor probability n={}, k={}, p={}, q={} is {}".format(i[0], i[1], i[2], i[3], '%.30f' % test_theor_2(i[0], i[1], i[2], i[3])))
    print("             Empiric probability is {}".format('%.30f' % test_empir_2(i[0], i[1], i[2], i[3])))
    print("---- ---- ---- ---- ---- ---- ---- ---- ---- ----")

```
