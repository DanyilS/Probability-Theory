```python
ans_ = nbinom.rvs(n = 5,p=0.4, size=100)
print(ans_)
```

```python
def f(arr, x):
    while (x in arr):
        arr.remove(x)
lst_r = sorted(list(ans_))
for i in lst_r:
    tmp = lst_r.count(i)
    f(lst_r,i)
    print("{} = {}".format(i, tmp))
```

```python
plt.hist(ans_)
plt.show()
```

```python
from scipy.stats import binom
```

```python
tst = [[15, 0.8], [15, 0.2], [20, 0.4], [20, 0.6], [7, 0.5], [25, 0.5], [100, 0.5]]
```

```python
for n, i in tst:
    x = np.arange(0, n+1)
    ans1 = binom.pmf(x, n, i)
    plt.subplot(1, 3, 1)
    plt.title('n = {}, p = {}'.format(n, i))
    plt.stem(x, ans1)
    ans2 = binom.cdf(x, n, i)
    plt.subplot(1, 3, 2)
    plt.plot(x, ans2, drawstyle='steps-post')
    ans3 = binom.rvs(n, i, size=10000)
    plt.subplot(1, 3, 3)
    plt.hist(ans3)
    plt.show()
```
