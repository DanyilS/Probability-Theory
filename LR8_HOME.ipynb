```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy.special import gammainc
from scipy.stats import norm, uniform
```

```python
#Task 1.1
k1, k2 = expon(), expon() # незалежні випадкові величини
r1, r2 = k1.rvs(10000), k2.rvs(10000) # генеруємо 10000 випадк значень величин
x = np.arange(0, 1, 0.01)

plt.hist(r1 / (r1 + r2), bins=60, density=True, rwidth=0.6, color='g')
plt.plot(x, [1]*len(x), 'r')
plt.show()
```
