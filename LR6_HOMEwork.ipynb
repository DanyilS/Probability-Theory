```python
# Tasl 1
import scipy.stats as stat
import numpy as np
def exp1(x):
    return x.mean()
def exp2(x):
    return x.expect()
def exp3(x):
    return x.stats()[0]
def dis(xi):
    def func1(x):
        return (x - xi.expect()) ** 2
    def func2(x):
        return x ** 2
    return [xi.var(), xi.expect(func1), xi.expect(func2) - xi.expect() ** 2, xi.stats()[1]]

distributions = [
                     ('Binom', stat.binom(n=10, p=0.4), 10 * 0.4, 10 * 0.4 * 0.6), 
                     ('geom', stat.geom(p=0.5), 1/0.5, 1 / 0.5),
                     ('pois', stat.poisson(0.6), 0.6, 0.6),
                     ('nbinom', stat.nbinom(n=5, p=0.4), 5*0.6 / 0.4, 0.6 * 5 / 0.4**2),
                     ('hgeom', stat.hypergeom(M=20, n=7, N=12), 7*12/20, 7*12*13*8 / 20**2 / 19),
                     ('uniform', stat.uniform(loc=0, scale=1), 1 / 2, 1 / 12), 
                     ('exponential', stat.expon(loc=0, scale=1/0.5), 1/0.5, 1/0.5**2),
                     ('normal', stat.norm(0, 1), 0, 1),
                     ('gamma', stat.gamma(2), 2, 2), 
                     ('lognorm', stat.lognorm(s=1, loc=0, scale=1), np.exp(0.5), (np.exp(1) - 1) * np.exp(1)), 
                     ('beta', stat.beta(2, 1.1), 2 / 3.1, 2.2 / (3.1**2*4.1))
                ]

for name, xi, m_t, v_t in distributions:
    print('{}\t\tExp = {}\tDis = {})'.format(name, m_t, v_t))
    e1 = exp1(xi)
    e2 = exp2(xi)
    e3 = exp3(xi)
    v = dis(xi)
    print( 'mean,expect,stats',e1,e2,e3, 'diff: {}'.format(  m_t-e1))
    print('Variation:', *v,  'diff: {}\n'.format( v_t-v[0]))
```

```python
# Task 5
import itertools
from collections import defaultdict
def comb(n, k):
    if n < k:
        return 0
    return int(np.math.factorial(n) / np.math.factorial(n - k) / np.math.factorial(k))
def pov_pereb(n):
    Exp, Dis = 0, 0
    om = range(1, 13)
    choices = list(itertools.product(om, repeat=n))
    md = defaultdict(int)
    for choice in choices:
        p = 0
        for el in set(choice):
            p += comb(choice.count(el), 2)
        md[p] += 1
    for k, v in md.items():    
        Exp += k * (v / len(choices))
        
    for k, v in md.items():    
        Dis += k ** 2 * (v / len(choices)) 
    Dis -= Exp ** 2
    
    return Exp, Dis


for n in range(1,7):
    print("When n = {}:\nExp = {}\nDis = {}\n".format(n, pov_pereb(n)[0], pov_pereb(n)[1] ))
```
