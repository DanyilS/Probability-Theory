```python
# Task 1.1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
def function(y):
    return np.arcsin(y / a) + np.pi / 2
a = 1/2 #шуканий параметр
stats = uniform(-0.5, 1)
 
plt.plot(np.arange(0, np.pi + 0.1, 0.1), a * np.sin(np.arange(0, np.pi + 0.1, 0.1)), 'r')
plt.hist(function(stats.rvs(size=100000)), density=True, rwidth=0.4, bins=60, color='g') # генеруємо 100000
plt.show()
```

```python
# Task 1.2
from scipy.stats import norm
def function1(data):
    l = list()
    for i in data:
        if not np.isnan(i):
            l.append(i)
    return l
def function2(x):
    return (-2*np.log((2*np.pi)**(1/2)*x))**(1/2)

data = function1(list(map(function2, norm.rvs(size=100000))))
plt.plot(np.arange(-3, 3, 0.1), (2*np.pi)**(-0.5)*np.exp(-(x**2)/2), 'r')
plt.hist([data,-np.array(data)],color = ['g','g'], density=True, rwidth=0.4, bins=60)
plt.show()
```
